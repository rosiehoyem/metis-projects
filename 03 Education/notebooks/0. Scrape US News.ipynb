```python
import re
import string

url = "http://www.usnews.com/education/best-high-schools/search?page="

def get_content(driver, url):
    driver.get(url)
    content = (driver.page_source).encode('ascii', 'ignore') 
    return content

def create_soup(content):
    soup = BeautifulSoup(content,"lxml")
    return soup

def iterate_through_urls(driver, url, data):
    for i in range(1,135):
        page_url = url + str(i)
        content = get_content(driver, page_url)
        soup = create_soup(content)
        scrape(soup, data)
    return data

def scrape(soup,data):
    divs = soup.find_all('div', class_='search-match-card-view')
    div = soup.find_all('div', class_='search-match-card-view')
    for div in divs:
        new_record = {}
        school_name = div.find('h3').text.strip()
        print("Scraping " + school_name)
        new_record['school_name'] = school_name
        cd = div.find('p').text.strip()
        new_record['city'] = cd.split(" | ")[0].split(",")[0].strip()
        new_record['state'] = cd.split(" | ")[0].split(",")[1].strip()
        new_record['district'] = cd.split(" | ")[1].strip()
        new_record['rank'] = div.find_all('span', class_='text-normal')[0].text.split()[0].strip()
        new_record['graduation_rate'] = div.find_all('span', class_='text-normal')[1].text.strip()
        new_record['college_readiness'] = div.find_all('span', class_='text-normal')[2].text.strip()
        new_record['medal_award'] = div.find_all('span', class_='text-normal')[3].text.strip()
        data[school_name] = new_record
    return data
```

```python
chromedriver = "./chromedriver"
driver = webdriver.Chrome(chromedriver)
data = {}
iterate_through_urls(driver, url, data)
```
