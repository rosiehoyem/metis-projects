```python
import pandas as pd
df = pd.read_csv('turnstile_160528.txt')
```

```python
df['countIn'] = df.ENTRIES.diff()
```

```python
df.columns = [u'C/A', u'UNIT', u'SCP', u'STATION', u'LINENAME', u'DIVISION', u'DATE',
       u'TIME', u'DESC', u'ENTRIES',
       u'EXITS',
       u'countIn']
```

```python
df['countOut'] = df.EXITS.diff()
```

```python
df['machine_change'] = (df.SCP != df.SCP.shift())
```

```python
df.head(50)
```

```python
df.countIn[df.index[df.machine_change==True]] = 0
```

```python
df.countOut[df.index[df.machine_change==True]] = 0
```

```python
df
```

```python
df['countSum'] = df.countIn + df.countOut
```
